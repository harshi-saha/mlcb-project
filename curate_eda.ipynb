```python
import pandas as pd
import numpy as np
```

```python
lincs_1000_path = "./curated/LINCS-Pilot1/L1000/replicate_level_l1k.parquet"
lincs_cp_path = "./curated/LINCS-Pilot1/CellPainting/replicate_level_cp_augmented.parquet"

bray_l1000_path = "./curated/CDRP-BBBC047-Bray/L1000/replicate_level_l1k.parquet"
bray_cp_path = "./curated/CDRP-BBBC047-Bray/CellPainting/replicate_level_cp_augmented.parquet"

caicedo_l1000_path = "./curated/LUAD-BBBC041-Caicedo/L1000/replicate_level_l1k.parquet"
caicedo_cp_path = "./curated/LUAD-BBBC041-Caicedo/CellPainting/replicate_level_cp_augmented.parquet"

rohban_l1000_path = "./curated/TA-ORF-BBBC037-Rohban/L1000/replicate_level_l1k.parquet"
rohban_cp_path = "./curated/TA-ORF-BBBC037-Rohban/CellPainting/replicate_level_cp_augmented.parquet"
```

```python
l1000_genes = pd.read_excel("./idmap.xlsx")
l1000_map = {k:v for k,v in zip(l1000_genes["probe_id"], l1000_genes["symbol"])}
```

```python
for i in [lincs_1000_path, bray_l1000_path, caicedo_l1000_path, rohban_l1000_path]:
    curr_df = pd.read_parquet(i)
    curr_df = curr_df.rename(columns=l1000_map)
    curr_df.to_parquet(i)
```

```python
lincs_l1000 = pd.read_parquet(lincs_1000_path)
lincs_cp = pd.read_parquet(lincs_cp_path)

bray_l1000 = pd.read_parquet(bray_l1000_path)
bray_cp = pd.read_parquet(bray_cp_path)

caicedo_l1000 = pd.read_parquet(caicedo_l1000_path)
caicedo_cp = pd.read_parquet(caicedo_cp_path)

rohban_l1000 = pd.read_parquet(rohban_l1000_path)
rohban_cp = pd.read_parquet(rohban_cp_path)
```
